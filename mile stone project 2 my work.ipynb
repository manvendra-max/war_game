```python
import random
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    def __init__(self):
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                created_card = Card(suit,rank)
                self.all_cards.append(created_card)
        
    def shuffle(self):
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player:
    def __init__(self,name):
        self.name = name
        self.all_cards = []
        
    def add_cards(self,new_cards):
        if type(new_cards)== type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
            
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def __str__(self):
        return f'player {self.name} has {len(self.all_cards)} cards.'
```

```python
## I created this function COMPARE to compare btw the cards on the table, or war in situation this will take out 6-6 cards from each player and will compare the last card of both  

def compare(player_one_cards, player_two_cards):
    print(len(player_one_cards))
    print('################')
    print(len(player_two_cards))
### if player one has more cards or both have 7 cards and player one has bigger value card
    if len(player_one_cards)>len(player_two_cards) or (len(player_one_cards)==len(player_two_cards) and player_one_cards[-1].value > player_two_cards[-1].value):    
        player_one.add_cards(player_one_cards)
        player_one.add_cards(player_two_cards)
        print('player one gets this hand')
        
    elif len(player_two_cards)>len(player_one_cards) or (len(player_one_cards)==len(player_two_cards) and player_one_cards[-1].value < player_two_cards[-1].value):   
        player_two.add_cards(player_two_cards)
        player_two.add_cards(player_one_cards)
        print('player two gets this hand')
```

```python
game_on = True
round_num = 0
while game_on:
    print(len(player_one.all_cards))
    print(len(player_two.all_cards))

    round_num += 1
    print(f'round num {round_num}')

    if len(player_one.all_cards) == 0:
        print('player One has been out of cards! \nplayer Two is winner!!!')
        game_on = False
        break
        
    if len(player_two.all_cards) == 0:
        print('player Two has been out of cards! \nplayer One is winner!!!')
        game_on = False
        break
        
    #game starts
    # Start a new round and reset current cards "on the table"
    player_one_cards = []
    player_one_cards.append(player_one.remove_one())
    
    player_two_cards = []
    player_two_cards.append(player_two.remove_one())
    
    if player_one_cards[0].value == player_two_cards[0].value:
        
        print("**************WAR**************")
        
        for i in range(6):
            if len(player_one.all_cards) > 0:
                player_one_cards.append(player_one.remove_one())
            if len(player_two.all_cards) > 0:
                player_two_cards.append(player_two.remove_one())
        compare(player_one_cards, player_two_cards)
        
    else:
        compare(player_one_cards, player_two_cards)
        
```
